```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
data = pd.read_csv("/mnt/c/Users/Vasiliki/Desktop/Stats/mouse_cognition(1).csv")
data.head(30)
```

```python

print("Shape:", data.shape)

df = data[['CREB1', 'object_recognition']].dropna()

print("N used in analysis:", len(df))
df.describe()
```

```python
sns.set_theme(style="whitegrid")

sns.regplot(data=df, x="CREB1", y="object_recognition")
plt.title("CREB1 vs Object Recognition")
plt.xlabel("CREB1")
plt.ylabel("Object recognition")
plt.show()
```

```python
from scipy import stats

r, p = stats.pearsonr(df["CREB1"], df["object_recognition"])

print("r =", r)
print("p =", p)
```

A Pearson correlation examined the relationship between CREB1 expression and object recognition in mice (N = 120). There was a strong positive correlation, r(118) = .855, p < .001, indicating that higher CREB1 expression is associated with better object recognition performance.
Biologically, this is consistent with CREB1’s role in synaptic plasticity and memory-related processes.
