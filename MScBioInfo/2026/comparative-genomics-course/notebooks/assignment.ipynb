# Assignment — Do Convergent Sites Cluster in the Transmembrane Domain?

## Background

In Notebook 3 you identified positions in the prestin alignment where
echolocating species show convergent amino acid usage. The results
are saved in `data/prestin_convergence_results.csv`.

Prestin has three major domains:
- **TMD** (transmembrane motor domain): approximately residues 81–505
- **Linker**: residues 506–530
- **STAS** (regulatory domain): residues 531–744

The TMD is responsible for prestin's electromotility — the motor
function that makes outer hair cells amplify sound. If convergent
evolution in echolocating species is driven by selection on the motor
function, we'd expect convergent sites to be **enriched in the TMD**
rather than spread randomly across the protein.

## Your task

**Create a new Jupyter notebook** that answers this question:

> Are convergent sites enriched in the transmembrane domain?

Use a **permutation test** (as you learned in Notebook 3) to test
whether the number of convergent sites in the TMD is greater than
expected by chance.

## What to include

1. **Load** the convergence results from Notebook 3
2. **Define** the domain boundaries (TMD: 81–505, STAS: 531–744)
3. **Count** how many convergent sites fall in each domain
4. **Permutation test**: shuffle the positions of convergent sites
   (pick the same number randomly from all positions) and count how
   many land in the TMD. Repeat 10,000 times.
5. **Plot** the null distribution with the observed value marked
6. **Interpret** the result in 2–3 sentences
